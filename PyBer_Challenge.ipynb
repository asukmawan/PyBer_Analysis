### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type - use groupby() function to create a Series that indexes the type of city, 
# and count ride_id
count_of_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
count_of_rides
```

```python
# 2. Get the total drivers for each city type - use the groupby() function to create a Series of data that has the type of city as the index, 
# then apply the sum() method to the "driver_count" column.
# used city_data_df as the merged pyber_data_df duplicates the driver_count due to the fact the ride_data is "left" and has multiple instances of the same city. 
# the starter data should have had "right" or merged the ride_data into the city_data, since city_data only has unique values for city
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
#  3. Get the total amount of fares for each city type - use the groupby() function to create a Series of data that has the type of city as the index, 
# then apply the sum() method to the "fare" column
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
#  4. Get the average fare per ride for each city type. - calculate the average fare per ride by city type by 
# dividing the sum of all the fares by the total rides
average_fare_per_ride = total_fares/count_of_rides
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. - calculate the average fare per driver by city type by 
# dividing the sum of all the fares by the total drivers
average_fare_per_driver = total_fares/total_drivers
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. - create a PyBer summary DataFrame with all the data gathered from 
# Steps 1-5, using the column names shown below
pyber_summary_df = pd.DataFrame({
    "Total Rides":count_of_rides, 
    "Total Drivers":total_drivers,
    "Total Fares":total_fares,
    "Average Fare per Ride":average_fare_per_ride,
    "Average Fare per Driver":average_fare_per_driver})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
# pyber_summary_df = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df = pyber_summary_df.style.format({
    "Total Rides":"{:,}",
    "Total Drivers":"{:,}",
    "Total Fares":"${:,.2f}",
    "Average Fare per Ride":"${:.2f}",
    "Average Fare per Driver":"${:.2f}"})
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
df = pyber_data_df.groupby(["date","type"]).sum()["fare"]
pd.DataFrame(df)
df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
df = df.reset_index()
df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
df_pivot= df.pivot("date", "type", "fare")
df_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
df_pivot_4_months = df_pivot.loc["2019-01-01":"2019-04-29"]
df_pivot_4_months
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
df_pivot_4_months.index = pd.to_datetime(df_pivot_4_months.index)
df_pivot_4_months
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
df_pivot_4_months.info()
# note that if the index just shows index it is an object
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
df_pivot_4_months_resampled = df_pivot_4_months.resample("W").sum()
df_pivot_4_months_resampled
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
import matplotlib.dates as mdates
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')


x_labels = ["Rural","Suburban", "Urban"]
fig,ax = plt.subplots(figsize=(20,6))
ax.set_title('Total Fare by City Type',fontsize=20)
ax.set_ylabel('Fare($USD)',fontsize=14)
ax.set_xlabel("2019")
ax.plot(df_pivot_4_months_resampled, label= x_labels)

formatter = mdates.DateFormatter("%b")
ax.xaxis.set_major_formatter(formatter)

locator = mdates.MonthLocator()
ax.xaxis.set_major_locator(locator)


ax.legend(x_labels, title="type", loc = "center", fontsize = "12", mode="Expanded")

# Save the figure.
plt.savefig("analysis/PyBer_fare_summary.png")

plt.show()
```
